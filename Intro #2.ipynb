```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
```

```python
position_info = pd.read_csv('data/positions.csv').iloc[0]
position_info
```

```python
position_opening_time = position_info['opening_time']
volume = position_info['volume']
ticker = position_info['ticker_yf']
```

```python
def get_current_ticker_price(ticker: str):

    today = pd.Timestamp.now(tz='Europe/Warsaw').floor('D')

    history = yf.Ticker(ticker).history(
        start=today + timedelta(days=-3), 
        end=today,
        interval='5m'
    )

    cond = (history.index == pd.Timestamp.now(tz='Europe/Warsaw').floor('5min'))

    if cond.sum() > 0: current_price = history[cond].iloc[0]
    else: current_price = history.iloc[-1]

    current_price = 0.5*(current_price['Open'] + current_price['Close'])

    return current_price
```
